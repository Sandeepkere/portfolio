```python

from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import re
import os

path = "/content/drive/MyDrive/datasets/liar_dataset/test.tsv"

df = pd.read_csv(path, sep='\t', header=None)

df.columns = ["jsonfile","label", "content", "topic", "speaker", "profession", "state", "party",
              "barely_true_total", "false_total", "half_true_total",
              "mostly_true_total", "pants_on_fire_total", "context"]

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^a-zA-Z0-9\s']", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["cleaned_content"] = df["content"].apply(clean_text)

def simplify_label(label):
    return 1 if label in ["true", "mostly-true"] else 0

df["binary_label"] = df["label"].apply(simplify_label)

df_cleaned = df[["cleaned_content", "binary_label"]].rename(
    columns={"cleaned_content": "text", "binary_label": "label"}
)

df_cleaned.head()

output_path = "/content/drive/MyDrive/misinfo_project/liar_dataset/liar_test.csv"
os.makedirs(os.path.dirname(output_path), exist_ok=True)

df_cleaned.to_csv(output_path, index=False)
```
