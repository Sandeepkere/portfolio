```python
import pandas as pd
import re


path = "/content/drive/MyDrive/datasets/liar_dataset/train.tsv"

df = pd.read_csv(path, sep='\t', header=None)

df.columns = ["jsonfile","label", "content", "topic", "speaker", "profession", "state", "party",
              "barely_true_total", "false_total", "half_true_total",
              "mostly_true_total", "pants_on_fire_total", "context"]

df.head()
```

```python
def clean_text_basic(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^a-zA-Z0-9\s']", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
df["cleaned_content"] = df["content"].apply(clean_text_basic)

def simplify_label(label):
    return 1 if label in ["true", "mostly-true"] else 0

df["binary_label"] = df["label"].apply(simplify_label)

df = df[df["cleaned_content"] != ""]
df = df.dropna(subset=["cleaned_content", "binary_label"])
df = df.drop_duplicates(subset=["cleaned_content"])


df.to_csv("cleaned_liar.csv", index=False)

df[["cleaned_content", "label", "binary_label"]].head()
```
